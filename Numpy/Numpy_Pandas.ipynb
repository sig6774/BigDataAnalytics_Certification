# Numpy

```python
import numpy as np
```

```python
# numpy 배열 생성
v1 = np.array([1,2,3,4])
print(v1)

# 연속되거나 일정한 규칙을 가진 배열 생성
v1 = np.arange(5)
# 0~4까지 값  출력 
print(v1)

v2 = np.arange(1, 10, 2, dtype= int)
print(v2)
# 간격을 가질 수 있게 만들 수 있음

v3 = np.arange(1, 10, 2) **2
print(v3)
# 간격에 제곱을 취해서 배열을 만들 수 있음
```

```python
# 행렬 만들기 
v1 = np.arange(12)
print(v1)

v2 = v1.reshape(2,6)
# 2행 6열의 행렬 생성
print(v2)

# 행렬 계산 
v1 = np.arange(4).reshape(2,2)

print(np.add(v1, v1)) # 덧셈
print(np.subtract(v1, v1)) # 뺄셈 
print(np.multiply(v1, v1)) # 곱셈 
print(np.dot(v1,v1)) # 행렬연산

# 다차원 행렬 
v1 = np.arange(12).reshape(2,3,2)
print("다차원 행렬 : \n" , v1)
```

# 데이터 변환 

```python
# 데이터 프레임 행 복사 
df_copy = df_csv.copy()
print(df_copy)
```

```python
# 데이터 프레임의 컬럼 이름 확인 
df_csv.columns
```

```python
# 데이터 컬럼 이름 수정 
print('변경 전 컬럼이름 : ', df_csv.columns)
df_csv.rename(columns ={'period' : 'time'}, inplace = True)
print('변경 후 컬럼이름 : ',df_csv.columns)
# columns를 통해 컬럼이름이 period에서 time으로 변경됨 
```

```python
# 데이터 프레임에 있는 컬럼들을 계산해서 새로운 변수 생성
df_csv['growth'] = df_csv['size'] / df_csv['time']
print(df_csv.head())
```

```python
# 특정 컬럼 삭제 
del df_csv['growth']
print(df_csv.head())
```

```python
# 데이터 케이스 추출 
df_brand = df_csv[(df_csv['continent'] == 1) & (df_csv['brand'] == 1)]
# continent열의 값이 1인 것과 brand 열의 값이 1인 데이터만 추출 
print(df_brand.head())
```

```python
# 데이터 케이스 추출 
df_over_size = df_csv[(df_csv['size'] >= 10) & (df_csv['time'] >= 30)]
# size열의 값이 10이상인 것과 time열의 값이 30 이상인 값 모두 추출 
print(df_over_size.head())
```

```python
# 데이터 값 변경 
# brand 열에는 1, 2, 3이 있는데 이것을 1과 2는 1, 3은 2로 변경 

df_csv['brand'].value_counts()
```

```python
# 방법 1
recode_brand = {'brand' : {1: 1, 2: 1, 3: 2}}
df_recode = df_csv.replace(recode_brand)
# 지정한조건을 데이터 프레임에 적용 
print(df_recode.head())
print("===========================")
print(df_recode["brand"].value_counts())
```

```python
# 방법 2 함수 생성
def brand_group(series):
    if series == 1:
        return 1
    elif series == 2:
        return 1
    else:
        return 2

df_csv['re_brand'] = df_csv['brand'].apply(brand_group)
print(df_csv.head())
print("===========================")
print(df_csv["re_brand"].value_counts())
```
