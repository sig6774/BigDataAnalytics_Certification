### 로지스틱 회귀모델 
- 종속 변수가 범주형 자료일 경우 적용하는 회귀모델 
- 로지스틱 회귀모델은 원자료를 확률 -> 오즈 -> 로그로 변환하여 선형회귀 모델을 적용 

```python
import pandas as pd 
data = pd.read_csv("../data/breast-cancer-wisconsin.csv", encoding="utf-8")
data.head()

X = data.iloc[:, 1:10]
y = data[['Class']]
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import *
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=410)
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.preprocessing import * 
help(sklearn.preprocessing.MinMaxScaler)
```

```python
scaler = MinMaxScaler()
# 0~1사이의 값으로 변환
scaler.fit(X_train)
X_scale_train = scaler.transform(X_train)
X_scale_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import * 
model = LogisticRegression() 
model.fit(X_scale_train, y_train)
pred_train = model.predict(X_scale_train)
model.score(X_scale_train, y_train)
```

```python
from sklearn.metrics import * 
help(sklearn.metrics.confusion_matrix)
```

```python
con = confusion_matrix(y_train, pred_train)
print(con)
```

```python
report_train = classification_report(y_train, pred_train)
print(report_train)
```

```python
pred_test = model.predict(X_scale_test)
model.score(X_scale_test, y_test)
```

```python
con_test = confusion_matrix(y_test, pred_test)
print(con_test)
```

```python
report_test = classification_report(y_test, pred_test)
print(report_test)
```

```python
from sklearn.model_selection import *
param_grid = {'C' : [0.001, 0.01, 0.1, 1, 10, 100]}
help(GridSearchCV)
```

```python
grid_search = GridSearchCV(LogisticRegression(), param_grid, cv = 5)
grid_search.fit(X_scale_train, y_train)
```

```python
print("Best Param : {}".format(grid_search.best_params_))
print("Best score : {}".format(grid_search.best_score_))
print("Test Score {}".format(grid_search.score(X_scale_test, y_test)))
```

```python
from scipy.stats import randint 
pram_dist = {"C" : randint(low = 0.001, high=100)}
random_search = RandomizedSearchCV(LogisticRegression(), pram_dist, n_iter = 100, cv = 5)
random_search.fit(X_scale_train, y_train)
```

```python
print("Best parameter : {}".format(random_search.best_params_))
print("Best score : {}".format(random_search.best_score_))
print("Test score : {}".format(random_search.score(X_scale_test, y_test)))
```
