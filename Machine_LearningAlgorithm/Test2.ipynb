- insurance 데이터 셋으로 charges 항목의 이상값의 합을 구하시오 

```python
import pandas as pd 
data = pd.read_csv("../Data/insurance.csv", encoding="utf-8")
```

```python
data.head()
```

```python
charges_std = data['charges'].std()
print(charges_std)
```

```python
standard1 = data['charges'].mean() + charges_std * 1.5
standard2 = data['charges'].mean() - charges_std * 1.5
print(standard1)
print(standard2)
```

```python
sum = 0
for i in data['charges']:
    if i > standard1 or i < standard2:
        sum += i
print(sum)
```

- E-commerce Shipping data의 train set을 참조하여 고객이 주문한 물품의 정시 도착 여부를 예측하시오

```python
import pandas as pd 
test = pd.read_csv("../Data/commerce/X_test.csv", encoding="utf-8")
X_train = pd.read_csv("../Data/commerce/X_train.csv", encoding="utf-8")
y_train = pd.read_csv("../Data/commerce/y_train.csv", encoding="utf-8")
```

```python
X_train.head()
```

```python
for i in X_train:
    print(i, "의 고유값들은 : " ,X_train[i].unique())
```

```python
# category : Warehouse_block, Mode_of_Shipment, Product_importance, Gender
# 나머지 : Customer_care_calls, Customer_rating, Cost_of_the_Product, Prior_purchases, Discount_offered, Weight_in_gms

X_category = X_train[['Warehouse_block', 'Mode_of_Shipment', 'Product_importance', 'Gender']]
X_category
```

```python
X_number = X_train[['Customer_care_calls', 'Customer_rating', 'Cost_of_the_Product', 'Prior_purchases', 'Discount_offered', 'Weight_in_gms']]
X_number
```

```python
X_category_dum = pd.get_dummies(X_category)
X_category_dum
```

```python
test_category = test[['Warehouse_block', 'Mode_of_Shipment', 'Product_importance', 'Gender']]
test_category
```

```python
test_number = test[['Customer_care_calls', 'Customer_rating', 'Cost_of_the_Product', 'Prior_purchases', 'Discount_offered', 'Weight_in_gms']]
test_number
```

```python
test_category_dum = pd.get_dummies(test_category)
test_category_dum
```

```python
# 이거는 몰랐던 거 
X_category_dum, test_category_dum = X_category_dum.align(test_category_dum, join='inner', axis = 1)
```

```python
from sklearn.preprocessing import * 

minmax = MinMaxScaler()
minmax.fit(X_number)
X_train_scaled = minmax.transform(X_number)
X_test_scaled = minmax.transform(test_number)

print(X_train_scaled)
```

```python
print(X_test_scaled)
```

```python
X_train_final = pd.concat([pd.DataFrame(X_train_scaled), X_category_dum], axis = 1)
# 데이터 합쳐줌 
X_test_final = pd.concat([pd.DataFrame(X_test_scaled), test_category_dum], axis = 1)
```

```python
X_train_final
```

```python
y_train = y_train[['Reached.on.Time_Y.N']]
y_train
```

```python
from sklearn.linear_model import * 
model = LogisticRegression()

model.fit(X_train_final, y_train)
```

```python
model.score(X_train_final, y_train)
```

```python
pred_test = model.predict_proba(X_test_final)
pred_test
```

```python
pred_test_prob = pd.DataFrame(pred_test[:,0], columns=['predict_prob'])
pred_test_prob
```

```python
final_predict = pd.concat([test['ID'], pred_test_prob], axis = 1)
```
