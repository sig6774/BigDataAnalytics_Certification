### Boston Housing 데이터셋이다. crim 항목의 상위에서 10번째 값으로 상위 10개의 값을 변환하고 age가 80이상인 값에 대하여 crim 평균을 구하시오

```python
import pandas as pd 
data = pd.read_csv("../Data/3rd_boston_housing.csv", encoding="utf-8")
data.head()
```

```python
help(pd.DataFrame.describe)
```

```python
print(data.sort_values(by='crim', ascending=False))
```

```python
data_sort = data.sort_values(by='crim', ascending=False)
```

```python
data_sort.head(10)
```

```python
standard = data_sort[9:10]['crim']
print(standard)
```

```python
def recreate(series):
    if series >= 25.9406:
        return 25.9406
    else:
        return series
```

```python
data_sort['crim'].apply(recreate)
```

```python
data_sort['re_crim'] = data_sort['crim'].apply(recreate)
print(data_sort.head(10))
```

```python
data_sort_age = data_sort[(data_sort['age'] >= 80)]
data_sort_age
```
