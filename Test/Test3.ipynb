- 국가별 입국자 수 데이터이며 2000년도 전체 입국자 평균보다 국내에 많이 입국한 국가의 수를 산출

```python
import pandas as pd 
data = pd.read_csv("../Data/3rd_tour.csv",encoding="euc-kr")
print(data)
```

```python
data2 = data[data['year'] == 2000]
# 2000년도 데이터만 추출
print(data2)
```

```python
# 평균 산출 
data3 = data2.iloc[:, 1:]
data3
```

```python
num = 0 
for i in data3:
    print(data3[i].values)
    num += data3[i].values
print(num)
```

```python
mean_con = int(num) / len(data3.columns)
```

```python
def country(series):
    if series >= mean_con:
        return series
    else:
        return "None"
```

```python
a = []
for i in data3:
    a.append(country(data3[i].values))

print(a)
```

```python
cnt = 0
for i in a:
    if i != "None":
        cnt+= 1
print(cnt)
# 야매로 풀었음
```

- 아래 여행객 보험 데이터 셋을 참조하여 여행객의 정보들을 기반으로 여행 보험상품 가입 여부를 예측(ID와 예측치를 csv 파일로 저장해서 제출)

```python
import pandas as pd 
test = pd.read_csv("../Data/3rd_TravelInsurancePrediction_test.csv", encoding="utf-8")
train = pd.read_csv("../Data/3rd_TravelInsurancePrediction_train.csv", encoding="utf-8")
```

```python
train.head()
```

```python
for i in train:
    print(i, " : " ,train[i].unique())
```

```python
# X, y 분리 
X_train = train[["Age", "Employment Type", "GraduateOrNot", "AnnualIncome", "FamilyMembers", "ChronicDiseases", "FrequentFlyer", "EverTravelledAbroad"]]
y_train = train["TravelInsurance"]

# category : Employment Type, GraduateOrNot, ChronicDiseases, FrequentFlyer, EverTravelledAbroad
# num : Age, AnnualIncome, FamilyMembers
```

```python
X_cat_train = X_train[['Employment Type', 'GraduateOrNot', 'ChronicDiseases', 'FrequentFlyer', 'EverTravelledAbroad']]
X_num_train = X_train[['Age', 'AnnualIncome', 'FamilyMembers']]
```

```python
X_cat_train_dum = pd.get_dummies(X_cat_train)
X_cat_train_dum
```

```python
test
```

```python
X_test = test[["Age", "Employment Type", "GraduateOrNot", "AnnualIncome", "FamilyMembers", "ChronicDiseases", "FrequentFlyer", "EverTravelledAbroad"]]
```

```python
X_cat_test = X_test[['Employment Type', 'GraduateOrNot', 'ChronicDiseases', 'FrequentFlyer', 'EverTravelledAbroad']]
X_num_test = X_test[['Age', 'AnnualIncome', 'FamilyMembers']]
```

```python
X_cat_test_dum = pd.get_dummies(X_cat_test)
X_cat_test_dum
```

```python
from sklearn.preprocessing import * 
minmax = MinMaxScaler()
minmax.fit(X_num_train)
X_scaled_train = minmax.transform(X_num_train)
X_scaled_test = minmax.transform(X_num_test)
```

```python
print(X_scaled_train, end="\n\n")
print(X_scaled_test)
```

```python
X_train = pd.concat([pd.DataFrame(X_scaled_train), X_cat_train_dum], axis=1)
X_train
```

```python
X_train_dum, X_test_dum =X_cat_train_dum.align(X_cat_test_dum, join="inner", axis = 1)
```

```python
X_cat_test_dum
```

```python
X_cat_train_dum
```

```python
X_test = pd.concat([pd.DataFrame(X_scaled_test), X_cat_test_dum], axis = 1)
X_test
```

```python
from sklearn.linear_model import * 
model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_train, y_train)
```

```python
pred_test = model.predict_proba(X_test)
pred_test
```
