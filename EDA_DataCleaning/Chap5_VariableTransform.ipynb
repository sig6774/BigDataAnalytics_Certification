## 변수 변환

### 로그 변환

```python
# 변수의 값을 로그로 변환하기 위해서는 numpy에서 log함수 사용
import numpy as np 
import pandas as pd 
data = pd.read_csv('Ex_CEOSalary.csv', encoding='utf-8')
```

```python
data['log_salary'] = np.log(data['salary'])
data['log_sales'] = np.log(data['sales'])
data['log_roe'] = np.log(data['roe'])
data.head
```

```python
data.hist(bins=50, figsize=(20,15))
# 로그 변환으로 이상치의 영향을 제거할 수 있음
```

```python
data.corr()
# 간단한 로그변환으로 원 데이터보다 상관계수가 높아진 것을 확인할 수 있음
```

### 제곱근 변환

```python
# 변수의 값을 제곱근으로 변화시키기 위해서 numpy의 sqrt 함수를 이용
data['sqrt_salary'] = np.sqrt(data['salary'])
data['sqrt_sales'] = np.sqrt(data['sales'])
data['sqrt_roe'] = np.sqrt(data['roe'])

data.head
```

```python
data.corr()
```
