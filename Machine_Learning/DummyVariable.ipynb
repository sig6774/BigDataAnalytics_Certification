### 범주형 변수 변환 
- 독립변수(X)에 범주형 변수가 있을 경우 원핫인코딩을 적용
- 원핫인코딩은 특점 범주만 표현하기 위해 적용할 수 있는 기법

```python
import pandas as pd 
data =  pd.read_csv('../Data/vote.csv', encoding = 'utf-8')
data.head()
```

```python
X1 = data[['gender', 'region']]

XY = data[['edu', 'income', 'age', 'score_gov', 'score_progress', 'score_intention', 'vote', 'parties']]
# 원핫인코딩을 할 대상을 나눔
```

```python
X1['gender'] = X1['gender'].replace([1,2], ['male', 'female'])
X1['region'] = X1['region'].replace([1,2,3,4,5], ['seoul', 'busan', 'incheon', 'ulsan', 'kyungki'])
# 숫자로 된 범주형 변수를 문자로 변환
```

```python
X1_dummy = pd.get_dummies(X1)
print(X1_dummy)
# 범주형 변수를 더미변수로 바꾸기 위해서 pd.get_dummies()함수 활용
```

```python
Fvote = pd.concat([X1_dummy, XY], axis = 1)
Fvote
```

```python
Fvote.to_csv('DummyVariables.csv', index=False, sep=',', encoding='utf-8')
```
