### 데이터 정규화 

```python
import pandas as pd 
data = pd.read_csv('DummyVariables.csv', encoding = "utf-8")
data.head()
```

```python
X = data.loc[:, "gender_female" : "score_intention"]
y = data[['vote']]
```

```python
from sklearn.model_selection import * 
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 410)
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.preprocessing import *
minmax = MinMaxScaler() 
minmax.fit(X_train)
```

```python
X_train_minmax = minmax.transform(X_train)
X_test_minmax = minmax.transform(X_test)
```

```python
pd.DataFrame(X_train_minmax).describe()
```

```python
from sklearn.linear_model import *
model = LogisticRegression()
```

```python
model.fit(X_train_minmax, y_train)
```

```python
pred_train = model.predict(X_train_minmax)
model.score(X_train_minmax, y_train)
```

```python
pred_test = model.predict(X_test_minmax)
model.score(X_test_minmax, y_test)
```

```python
from sklearn.metrics import *
confusion_train = confusion_matrix(y_train, pred_train)
print(confusion_train)
```

```python
confusion_test = confusion_matrix(y_test, pred_test)
print(confusion_test)
```
