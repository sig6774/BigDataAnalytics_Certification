```python
import warnings 
warnings.filterwarnings('ignore')
import pandas as pd 
data = pd.read_csv('DummyVariables.csv', encoding = "utf-8")
data.head()
```

### Split dataset

```python
X = data.loc[:,"gender_female" : "score_intention"]
y = data[['vote']]
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import *
X_train, X_test, y_train, y_test = train_test_split(X,y, stratify = y, random_state = 410)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
import sklearn
test = dir(sklearn)
for i in test:
    print(i)
```

```python
from sklearn.preprocessing import *
test = dir(sklearn.preprocessing)
for i in test:
    print(i)
```

```python
minmax = MinMaxScaler()
minmax.fit(X_train)
X_train_scale = minmax.transform(X_train)
X_test_scale = minmax.transform(X_test)
# 연습 
```

```python
from sklearn.linear_model import *
model = LogisticRegression() 

# 교차검증 
scores = cross_val_score(model, X_train, y_train, cv = 5)
print(scores)
print(scores.mean())
# 훈련데이터를 cv의 수만큼 나눠서 각각 훈련하고 결과를 도출 
# 다양하게 모델을 훈련시켜 과대, 과소 적합 예방
```

```python
test = dir(sklearn.model_selection)
for i in test:
    print(i)
```

```python
help(sklearn.model_selection.KFold)
```

```python
# cross validation은 데이터셋의 순서대로 구간을 나누며 만약 순서가 결과에 영향을 준다면 잘못된 결과가 나올 수 있음
# KFold는 데이터 셋을 나눌 때 무작위로 나눔
kfold = KFold(n_splits = 5, shuffle = True, random_state = 410)
score = cross_val_score(model, X_train, y_train, cv = kfold)
print(score)
```

```python
shuffle_split = ShuffleSplit(test_size=0.5, train_size=0.5, random_state = 410)
score = cross_val_score(model, X_train, y_train, cv = shuffle_split)
print(score)
```
